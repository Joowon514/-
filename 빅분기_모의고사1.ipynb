## 2유형

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')

display(train.head())
display(test.head())
```

```python
X_train = train.loc[:, 'ID':'poutcome']
y_train = train[['y']]
X_train.head()
```

```python
y_train.head()
```

```python
y_train['y'].value_counts()
```

```python
y_train['y'] = y_train['y'].replace({'yes', 'no'}, {0, 1})
```

```python
X_train.info()
```

```python
X_train_day = X_train.pop('day')
test_day = test.pop('day')
X_train_month = X_train.pop('month')
test_month = test.pop('month')

X_train_id = X_train.pop('ID')
test_id = test.pop('ID')
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

encoder.fit(X_train['job'])
X_train['job'] = encoder.transform(X_train['job'])
test['job'] = encoder.transform(test['job'])

encoder.fit(X_train['education'])
X_train['education'] = encoder.transform(X_train['education'])
test['education'] = encoder.transform(test['education'])

X_train = pd.get_dummies(X_train)
test = pd.get_dummies(test)

X_train.head()
```

```python
from sklearn.model_selection import train_test_split
X_tra, X_val, y_tra, y_val = train_test_split(X_train, y_train, stratify=y_train, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_tra, y_tra)
```

```python
pred_val = model.predict(X_val)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(pred_val, y_val)
```

```python
model.fit(X_train, y_train)
pred_train = model.predict(X_train)
roc_auc_score(pred_train, y_train)
```

```python
pred_test = model.predict(test)
```

```python
sub_1 = pd.DataFrame({'ID':test_id, 'predict': pred_test})
sub_1.to_csv('sub_1.csv', index=False)
```
